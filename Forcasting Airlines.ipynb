## Visualization

```python
data.Passengers.plot(figsize=(12,4))
```

```python
plt.scatter(data['Date'],data['Passengers'])
plt.xlabel('Date')
plt.ylabel('Passengers')
plt.title('Airline passenger traffic')
plt.show()
```

```python
## Lone plot

data['Passengers'].plot(figsize=(20,8),color='green', marker='o')
plt.show()
```

**Here we can see there is increasing trend going on.**
**Number of passengers are increasing with the time**

**Trend and Variance is not constant**

**Hence time series is not stationary**

```python
# Making Separate Month and Year columns

data_1 = data.copy()

data_1['Date'] = pd.to_datetime(data_1.Date,format='%b-%y')
data_1['Month'] = data_1.Date.dt.strftime('%b')              # Month Extraction
data_1['Year'] = data_1.Date.dt.strftime('%Y')              # Year Extraction

data_1.head()
```
